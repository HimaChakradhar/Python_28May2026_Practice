```python
def sum_of_digits(num):
  total= 0
  for digit in str(num):
    total= total+ int(digit)
  return total

result= sum_of_digits(12345)
print(result)

```

```python
def is_prime(num):
    if num<2:
        return False
    for i in range(2,num):
      if num % i == 0:
        return False
      else:
        return True
    
print(is_prime(16)) 
```

```python
def factorial(num):
  fact=1
  for i in range(1, num+1):
     fact = fact * i
  return fact
  
result = factorial(4)
print(result)

```

```python
def vowels_Count(inputstr):
  vowels = "aeiouAEIOU"
  count=0
  for ch in inputstr:
    if ch in vowels:
      #print(ch)
      count = count+1
  return count
    
result= vowels_Count("Tomoto")
print(result)
```

```python
def find_max_list(num):
  max = num[0]
  for i in num:
    if i> max:
     max=i
  return max
result = find_max_list([10,20,9,59,60,99])
print(result)
                  
  
```
