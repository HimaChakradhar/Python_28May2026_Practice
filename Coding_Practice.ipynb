```python
firstamount = int(input("Enter first amount"))
secondamount = int(input("Enter second amount"))
print(f"Enter the first amount : {firstamount}")
print(f"Enter the second amount: {secondamount}")
sum= firstamount + secondamount
differance = firstamount-secondamount
product = firstamount* secondamount
quotient=firstamount/secondamount
Remainder = firstamount % secondamount
print(f"Sum:{sum}")
print(f"Difference: {differance}")
print(f"Product: {product}")
print(f"Quotient:{quotient}")
print(f"Remainder:{Remainder} ")
```

```python
post = input("Enter your post ")
print(f"Enter your post: " + post)
upper = post.upper()
lower = post.lower()
reverse = post[::-1]
print(f"UpperCase:" +upper)
print(f"LowerCase:"+ lower)
print(len(post))
print(f"Reversed:"+ reverse)
```

```python
marks= int(input("Enter the student marks"))

if marks>=90 :
  print("Grade: A \n Well done!")
elif marks>=75:
  print("Grade: B \n Well done!")
elif marks>=60 :
   print("Grade: C")
elif marks>=50:
  print("Grade:D")
else:
   print("Fail")
```
